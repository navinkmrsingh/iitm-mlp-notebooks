```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Practice Assignments

```python
from sklearn.datasets import load_wine
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

## Q1 to 3

```python
X, y = load_wine(as_frame=True, return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
```

### Q1

```python
clf = DecisionTreeClassifier(random_state=1)
clf.fit(X_train, y_train).score(X_train, y_train)
```

### Q2

```python
clf.score(X_test, y_test)
```

### Q3

```python
from sklearn import tree

column_names = X.columns

plt.figure(figsize=(48,18), facecolor='w')

a = tree.plot_tree(
    clf,
    feature_names=column_names,
    rounded=True,
    filled=True,
    fontsize=12
)
plt.show()
```

```python
clf.tree_.max_depth
```

# Graded Assignment

## Q1 to 3

```python
X, y = load_wine(as_frame=True, return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

clf = DecisionTreeClassifier(random_state=1)
```

```python
from sklearn.model_selection import GridSearchCV

tune_parameters = [{
    'criterion': ['entropy', 'gini'],
    'splitter': ['random', 'best'],
    'min_samples_leaf': [2, 4, 6, 8, 10],
    'max_depth': [3, 4, 5, 6]
}]

    
clf_CV = GridSearchCV(
    DecisionTreeClassifier(), 
    tune_parameters,    
    cv=4
)

clf_CV.fit(X_train, y_train)
```

### Q1

```python
clf_CV.best_estimator_.score(X_test, y_test)
```

### Q2

```python
clf_CV.best_estimator_.tree_.max_depth
```

### Q3

```python
clf_CV.best_estimator_.min_samples_leaf
```
